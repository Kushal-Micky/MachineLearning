# Neural Networks

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = 1.5
y = 0.3
```

```python
W = np.random.randn(1)
print (W , W.shape)
```

```python
#Bulid the training loop
epochs = 20
lr = 0.1
eList = []
for epoch in range (epochs):
    y_out = X*W 

    #Compute the error
    error = np.square(y -y_out)
    eList.append (error) 

    #Cmpute the gradient
    gradient = -2 *(y-y_out)

    #BackPropagate the error 
    W = W - gradient*lr
    
```

```python
plt.plot (range(epochs),eList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs Vs Error")
```

# Neural Networks with many weights

```python
X = np.array([[1.5 , 0.625]])
y =0.5

W = np.random.randn(1,2)

epochs = 20
lr = 0.1
eList = []
for epoch in range (epochs):
    y_out = np.dot (X, np.transpose(W))
    #Compute the error
    error = np.square(y -y_out)
    eList.append (error[0][0]) 

    #Cmpute the gradient
    gradient = -2 *(y-y_out)

    #BackPropagate the error 
    W = W - gradient*lr
    
plt.plot (range(epochs),eList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs Vs Error")
```
