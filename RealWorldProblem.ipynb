```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
m=0.7
c=0.3

X = np.arange(0,1,0.02)
y = m*X + c

print ( X.shape, y.shape )
```

```python
split = int(0.8*len(X))
X_train, y_train = X[:split], y[:split]
X_test, y_test = X[split:] , y[split:]
```

```python
def plt_data (X_train = X_train , y_train = y_train, X_test = X_test, y_test = y_test, predictions = None):
    plt.figure (figsize=(8,6))
    plt.scatter(X_train, y_train, c='b', s = 8, label="Training Data")
    plt.scatter(X_test, y_test, c='r', s = 8, label="Testing Data")
    if predictions is not None:
        plt.scatter(X_test, predictions, c = 'g', s=8, label="Predicted Data")
    plt.legend()

plt_data()
```
